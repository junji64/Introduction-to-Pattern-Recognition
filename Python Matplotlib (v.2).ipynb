### Plotting
matplotlib의 가장 중요한 기능은 플롯(plot)으로 2D 데이터를 플로팅 할 수 있습니다. 다음은 간단한 예입니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

추가 작업을 조금만 하면 한 번에 여러 줄을 쉽게 그릴 수 있고, 제목, 범례 및 축 레이블을 추가 할 수 있습니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine using pyplot()')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

[여기](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)에서 plot 함수에 대해 더 많이 읽을 수 있습니다.
또는, 다음과 같이 subplots()을 사용하여 같은 플로팅을 할 수도 있다.

```python
fig, ax = plt.subplots()
ax.plot(x, y_sin, x, y_cos)
ax.set_title("Sine and Cosine using pyplot.subplots()")
ax.set_xlabel("x axis label")
ax.set_ylabel("y axis label")
ax.legend(['Sine', 'Cosine'])
plt.show()
```

### subplot
subplot 함수를 사용하여 같은 그림에 다른 것들을 그릴 수 있습니다. 다음은 그 예입니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

x = np.linspace(0, 5, 10)
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].grid()
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_xlim([2, 5])
axes[2].set_ylim([0, 60])
axes[2].set_title("custom axis range")
plt.show()
```

[여기](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplots)에서 subplots 함수에 대한 자세한 내용을 볼 수 있습니다.

### Bar

```python
import matplotlib.pyplot as plt

x = [2, 4, 6, 8, 10]
y = [6, 7, 8, 2, 4]

x2 = [1, 3, 5, 7, 9]
y2 = [7, 8, 2, 4, 2]

plt.bar(x, y, label="Bars 1", color="r")
plt.bar(x2, y2, label="Bars 2", color="g")

plt.show()
```

### Scatter plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]

plt.scatter(x, y, label="scatter plot", color="b", marker="*")
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.legend()
plt.show()
```

### Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 100
sigma = 15
x = mu + sigma * np.random.randn(437)

num_bins = 50

fig, axes = plt.subplots()

n, bins, patches = axes.hist(x, num_bins, density=1)
y = ((1 / (np.sqrt(2 * np.pi) * sigma )) *
     np.exp(-0.5 * (1 / sigma * (bins -mu))**2))
axes.plot(bins, y, '--')
axes.set_xlabel('Smarts')
axes.set_ylabel('Probability density')
axes.set_title('Histogram of IQ: $\mu = 100$, $\sigma = 15$')
             
fig.tight_layout()
plt.show()
```

### 이미지
imshow 함수를 사용하여 이미지를 표시 할 수 있습니다. 다음은 그 예입니다.

```python
import numpy as np

img = plt.imread('cat.jpg')
print(img.dtype, img.shape)

# We can tint the image by scaling each of the color channels
# by a different scalar constant. The image has shape (400, 248, 3);
# we multiply it by the array [1, 0.95, 0.9] of shape (3,);
# numpy broadcasting means that this leaves the red channel unchanged,
# and multiplies the green and blue channels by 0.95 and 0.9
# respectively.
img_tinted = img * [1, 0.95, 0.9]
print(img_tinted.dtype, img_tinted.shape)

plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(np.uint8(img_tinted))
plt.show()
```

### 포인트 간 거리
SciPy는 점 집합 사이의 거리를 계산하기 위한 몇 가지 유용한 함수를 정의합니다.
scipy.spatial.distance.pdist 함수는 주어진 집합에서 모든 점 쌍 사이의 거리를 계산합니다 :

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform

x = np.array([[0,1], [1,0], [2,0]])
print(x)
plt.scatter(x[:,0], x[:,1], color="b", marker="o")
plt.show()

d = squareform(pdist(x, 'euclidean'))
print(d)
```
